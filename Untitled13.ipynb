```python
import pandas as pd
import sqlite3
import matplotlib.pyplot as plt

# 1. Load the CSV file into a DataFrame
df = pd.read_csv('/content/sales_data_sample.csv', encoding='latin1')

# 2. Create an inâ€‘memory SQLite database and write the DataFrame to a table
conn = sqlite3.connect(':memory:')
df.to_sql('sales', conn, index=False, if_exists='replace')

# 3. Use SQL to get the basic sales summary
total_quantity = pd.read_sql(
    'SELECT SUM(QUANTITYORDERED) AS total_quantity FROM sales', conn
)['total_quantity'][0]

total_revenue = pd.read_sql(
    'SELECT SUM(QUANTITYORDERED * PRICEEACH) AS total_revenue FROM sales', conn
)['total_revenue'][0]

# 4. Print a quick textual summary
print(f"Total quantity sold: {total_quantity}")
print(f"Total revenue: ${total_revenue:,.2f}")

# 5. Visualise the two metrics with a very simple bar chart
metrics = ['Total Quantity', 'Total Revenue']
values = [total_quantity, total_revenue]

plt.bar(metrics, values)  # default colours, single plot
plt.ylabel('Value')
plt.title('Sales Summary')
plt.tight_layout()
plt.show()
```

```python
# Improve visuals for better readability and aesthetics
fig, ax = plt.subplots(figsize=(8, 6))

# Convert total revenue to millions for better scaling on the bar chart
values_scaled = [total_quantity, total_revenue / 1_000_000]
labels = ['Total Quantity Sold', 'Total Revenue (in Millions)']

bars = ax.bar(labels, values_scaled, color=['skyblue', 'mediumseagreen'])

# Add value labels on top of the bars
for bar in bars:
    yval = bar.get_height()
    ax.text(bar.get_x() + bar.get_width()/2, yval + 0.05, f'{yval:.2f}', ha='center', va='bottom', fontsize=10, fontweight='bold')

# Style the chart
ax.set_title('Sales Summary Overview', fontsize=16, fontweight='bold')
ax.set_ylabel('Values', fontsize=12)
ax.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
